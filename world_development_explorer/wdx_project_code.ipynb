```python
#I wanted to make plotly charts out of my pandas dataframe
import pandas as pd
```

```python
#the plotly charts are displayed using HTML from the ipython library
from IPython.display import HTML
import plotly.express as px

```

```python
df=pd.read_csv("/content/wdi_data (1).csv")
df #
```

# **CO2 EMISSIONS OF TOP ECONOMIES DURING 2010 TO 2018**

```python
fig = px.line(df, x="Year", y="EN.ATM.CO2E.KT", color="Country Name");
fig.show();
```

```python
df_ghg=pd.read_csv("/content/ghg-table.csv")
df_ghg
```

# **OTHER GREEN HOUSE GAS EMISSIONS OF TOP ECONOMIES DURING 2010 TO 2018**

```python
fig = px.choropleth(df_ghg, locations="Country Code",
                    color="EN.ATM.GHGO.KT.CE", 
                    hover_name="Country Name", 
                    color_continuous_scale=px.colors.sequential.Plasma)
fig.show()
```

# **THIS BUBBLE CHART SHOWS THAT TOP ECONOMIES ARE THE COUNTRIES THAT HAVE HIGHEST CARBON EMISSIONS AND ALSO HAVE EXCEEDED WHO'S AIR POLLUTION GUIDELINES IN THE TIME SPAN 2010 TO 2018**

```python
fig = px.scatter(df_bubble.query("Year==2014"), x="NY.GDP.PCAP.PP.CD", y="lifeExp",
	         size="pop", color="continent",
                 hover_name="country", log_x=True, size_max=60)
fig.show()
```
