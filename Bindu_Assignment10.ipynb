## Running notes



1. Wide data : A table stored in ‘wide’ form spreads a variable across several columns


2. Long data/tidy data : A table stored in ‘long’ form has a single column for each variable in the system.


3. Pandas melt() function: used to change the DataFrame format from wide to long. It's used to create a specific format of the DataFrame object where one or more columns work as identifiers. All the remaining columns are treated as values and unpivoted to the row axis and only two columns variable and value.


4. The split() method splits the string from the specified separator and returns a list object with string elements.


5. Pandas Series is a one-dimensional labeled array capable of holding any data type.


6. Pandas Series can be created out of a Python list or NumPy array. It has to be remembered that unlike Python lists, a Series will always contain data of the same type. This makes NumPy array a better candidate for creating a pandas series


7. Series & indexing:
it’s possible to have our own row index values while creating a Series.


8. if there are 5 elements present within a list. Then the first element (the leftmost element) holds the “zeroth” place "0 index", followed by the elements in first, second, third, and fourth positions.


9. index() method with argument as item name -->gives the index of the item. And indexing operator is [] in python. 


10. There is also negative indexing in python.(look into it later) for example item[-1].


11. loc[] method is a method that takes only index labels and returns row or dataframe if the index label exists in the caller data frame.


12. Python merge function: Combining Data on Common Columns or Indices.

13. to_csv : 
df.to_csv(r'Path where you want to store the exported CSV file\File Name.csv', index = False)
